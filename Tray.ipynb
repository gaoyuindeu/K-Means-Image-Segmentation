```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from sklearn.cluster import KMeans
from PIL import Image
from skimage.color import rgb2hsv
```

```python
img = Image.open(r'D:\Downloads\tray-food-segmentation\TrayDataset\TrayDataset\XTrain\1001a01.jpg')
```

```python
x = np.array(img)
```

```python
imshow(x)
```

```python
z = np.dstack((x,rgb2hsv(x)))
```

```python
z.shape
```

```python
vectorized = np.float32(z.reshape((-1,6)))
```

```python
kmeans = KMeans(random_state=0, init='random', n_clusters=6)
```

```python
james = kmeans.fit_predict(vectorized)
```

```python
pic = james.reshape(256,416)
```

```python
imshow(pic)
```

```python
f, axarr = plt.subplots(1,2,figsize=(15,15))
axarr[0].set_xlabel('Original Image', fontsize=20)
axarr[1].set_xlabel('Segmented Image', fontsize=20)  
axarr[0].imshow(x)
axarr[1].imshow(pic)
```
